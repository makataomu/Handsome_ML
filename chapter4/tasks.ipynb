```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def non_convex_function(x, y):
    return np.sin(5 * x) * np.cos(5 * y) / 5 + x**2 + y**2

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = non_convex_function(X, Y)

# 3D surface plot
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(121, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title('Non-Convex Function')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
fig.colorbar(surf, shrink=0.5, aspect=5)

# Contour plot
ax = fig.add_subplot(122)
contour = ax.contourf(X, Y, Z, levels=50, cmap='viridis')
ax.set_title('Non-Convex Function (Contour)')
ax.set_xlabel('x')
ax.set_ylabel('y')
fig.colorbar(contour)

plt.tight_layout()
plt.show()

# Demonstrating multiple local minima
from scipy.optimize import minimize

def find_local_minimum(start_point):
    result = minimize(lambda args: non_convex_function(*args), start_point, method='BFGS')
    return result.x, result.fun

# Starting from different initial points
start_points = [(-1.5, -1.5), (0, 0), (1.5, 1.5)]
for start in start_points:
    minimum_point, minimum_value = find_local_minimum(start)
    print(f"Starting from {start}:")
    print(f"Local minimum found at {minimum_point}")
    print(f"Function value at minimum: {minimum_value}")
    print()
```
